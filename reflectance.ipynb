```python
import numpy as np
import matplotlib.pyplot as plt

n_rows = 100
read_verif_file = lambda folder, i: np.fromfile(folder + '/verification_reflectance_' + str(i) + '.wsv', sep=' ').reshape(n_rows + 1)*1e-2
read_data_file = lambda folder: np.loadtxt(folder + '/reflectance.csv', delimiter=',')

read_folder = lambda folder: [read_data_file(folder)] + [read_verif_file(folder, i) for i in (1, 2)]

results = [read_folder(folder) for folder in ['one_layer_matched', 'one_layer_unmatched', 'three_layer']]
```

```python
data, verif_1, verif_2 = read_folder('one_layer_matched')

for data, verif_1, verif_2 in results:
    plt.figure()
    plt.plot(range(n_rows), data, label="data")
    plt.plot(range(n_rows), verif_1[:-1], label="verif_1")
    plt.plot(range(n_rows), verif_2[:-1], label="verif_2")
    plt.legend()
    plt.yscale('log')
    plt.xlim((0, 75))
    plt.show()
```
